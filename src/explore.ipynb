```python
#Removing duplicated columns (the "title one")
total_data = data_cinema.loc[:, ~data_cinema.columns.duplicated()]

total_data.head()
```

```python
total_data.to_csv("/workspaces/machine-learning-k-nearest-neighbors/data/raw/total_data.csv", index = False)
```

```python
#Removing columns
data=total_data.copy()
data.columns
```

```python
columns = ['movie_id', 'title', 'overview','genres', 'keywords', 'cast', 'crew']
data=data[columns]
data.head()
```

```python
#Transforming the data

import json
import pandas as pd

def extract_names(json_str):
    if pd.notna(json_str):
        try:
            names_list = [item["name"] for item in json.loads(json_str)]
            return ','.join(names_list)
        except json.JSONDecodeError:
            return ''
    else:
        return ''


data["genres"] = data["genres"].apply(extract_names)
data["keywords"] = data["keywords"].apply(extract_names)
data["cast"] = data["cast"].apply(extract_names)
data.head()
```

```python
def extract_3_names(text):
    return ','.join(text.split(',')[:3])

data["cast"] = data["cast"].apply(extract_3_names)
data.head()
```

```python
def getdirector(crew_json):
    crew_data = json.loads(crew_json)
    for crew in crew_data:
        if crew.get('job') == 'Director':
            return crew.get('name')
    return None

data['crew'].apply(getdirector)

data['crew'] = data['crew'].apply(getdirector)
data.head()
```

```python
data.info()
```

We found 30 null values in the crew column and 3 in the overview one, so we will be removing these entries from the dataset.

```python
data = data.dropna(subset=['crew', 'overview'])
```

```python
#Dropping spaces from names

data['genres'] = data['genres'].str.replace(' ','')
data['cast'] = data['cast'].str.replace(' ','')
data['crew'] = data['crew'].str.replace(' ','')
data['keywords'] = data['keywords'].str.replace(' ','')

data.head()
```

```python
import numpy as np

def split(column, delimiter, df):
    data_clean = data.copy()
    data_clean[column] = data_clean[column].apply(lambda x: x.split(delimiter) if isinstance(x, str) else np.nan)
    return data_clean


data= split('overview', ' ', data)

remaining_columns = ['keywords', 'genres', 'cast']
for c in remaining_columns:
    data= split(c, ',', data)


```

```python
data.head()
```

```python
#Creating column "tag"

new_df = data['overview'] + data['genres'] + data['keywords'] + data['cast']

data['tags']= new_df.apply(lambda lst: ' '.join(lst))
data['tags']= data['tags'] + ' ' + data['crew']

data.drop(columns = ["genres", "keywords", "cast", "crew", "overview"], inplace = True)

data.iloc[0].tags
```

```python
#Building a KNN
#Vectorizing

from sklearn.feature_extraction.text import CountVectorizer

vec_model = CountVectorizer(stop_words = "english")
data_vec = vec_model.fit_transform(data['tags']).toarray()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(data_vec)
```

```python
def recommend(movie):
    # Verifica si la película está presente en el DataFrame
    if movie in data["title"].values:
        movie_index = data[data["title"] == movie].index[0]
        distances = similarity[movie_index]
        movie_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x: x[1])[1:6]
        
        print(f" The best 5 recommendations if you like '{movie}':")
        for i, (index, distance) in enumerate(movie_list, start=1):
            recommended_movie = data.iloc[index].title
            print(f"{i}. {recommended_movie}")
         
recommend("John Carter")
```
